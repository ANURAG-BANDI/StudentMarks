```python
#Import rquired packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import the data
data = pd.read_excel("C:/Users/Anvesh Bandi/Desktop/DATASETS/GRIP_TASK#1.xlsx")
```

```python
data.head(15)
```

```python
#Dropout any missing values
data.dropna()
```

```python
data.describe()
```

```python
#plot the data
data.plot(x='Hours',y='Scores',style='s')
plt.title('Hours vs Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#Assign variables to X & y and also reshape the data to 2D
X = np.array(data['Hours']).reshape(-1,1)
y = np.array(data['Scores'])
```

```python
print(X)
```

```python
print(y)
```

```python
#Split the data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state=0)
```

```python
model = LinearRegression()

model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
#predicting the model values
y_predict = model.predict(X_test)
y_predict
```

```python
#Plotting test data
plt.scatter(X_test,y_test,color='r')
plt.plot(X_test, y_predict, color='g')
plt.title('plot for test data')
plt.xlabel('Hours studied')
plt.ylabel('Scored')
```

```python
#Compare predicted with Actual values
predicted_values = pd.DataFrame({'predicted':y_test, 'Actual':y_predict})
predicted_values
```

```python
#predicting the given question(predicted score for 9.5hours of study)
Answer = model.coef_*(9.25)+model.intercept_
print(Answer)
```

```python
#Measuring the accuracy
from sklearn.metrics import mean_absolute_error
```

```python
MAE = mean_absolute_error(y_test, y_predict)
MAE
```
